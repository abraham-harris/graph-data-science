## 2.

```python
# Complete graph
G = nx.complete_graph(6)
pu.show_graph(G)
```

```python
# Barbell graph
G = nx.barbell_graph(6, 4)
pu.show_graph(G)
```

```python
# Cycle graph
G = nx.cycle_graph(12)
pu.show_graph(G)
```

```python
# Circulant graph (like a cycle but not several nearest neighbors)
G2 = nx.circulant_graph(20, [1, 2, 3])
pu.show_graph(G2)
```

```python
# Grid graph (grids of arbitrary dimension)
G = nx.grid_graph((2,3,2))
pu.show_graph(G)
```

```python
# Ladder graph
G = nx.ladder_graph(7)
pu.show_graph(G)
```

## 4.

```python
# Barbell graph
G = nx.barbell_graph(3, 1)

# Set colors
color_map = [
    "salmon", "sandybrown", "khaki", "palegreen", "aquamarine", "plum", "lightpink"
]

# Set sizes
node_sizes = [150, 200, 250, 300, 350, 400, 450]

nx.draw(G, node_color=color_map, node_size=node_sizes, with_labels=True)
```
